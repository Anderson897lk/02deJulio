# Ejercicio 1


```
import numpy as np
import matplotlib.pyplot as plt
# 1
# Función dada
def f(x):
    return np.tan(x**2) - 1/2

# Método de bisección
def biseccion(a, b, tol=1e-4):
    print(f"f(a) = {f(a)}, f(b) = {f(b)}")  # Imprimir los valores en los extremos
    if f(a) * f(b) > 0:
        print("No hay cambio de signo en el intervalo, prueba otro intervalo.")
        return None

    # Aplicar el método de bisección
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c  # La raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Intervalo adecuado [1.5, 2] (donde la función cambia de signo)
raiz = biseccion(1.5, 2, tol=1e-4)
print(f"La raíz aproximada en el intervalo [1.5, 2] es: {raiz}")

# Graficar la función en el intervalo [0, 3]
x_vals = np.linspace(0, 3, 400)  # Genera puntos en el intervalo [0, 3]
y_vals = f(x_vals)  # Evalúa la función en esos puntos

# Graficar la función
plt.plot(x_vals, y_vals, label=r"$y = \tan(x^2) - \frac{1}{2}$")  # Corregido el formato de LaTeX
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal en y=0
plt.axvline(0, color='black',linewidth=1)  # Línea vertical en x=0
plt.title(r"Gráfica de $y = \tan(x^2) - \frac{1}{2}$")  # Corregido el título con LaTeX
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# Ejercicio 5

```
import numpy as np
import matplotlib.pyplot as plt
# 5
# Función dada
def f(x):
    return 2 * np.exp(x) - 7 * x**3 - 2

# Método de Regla Falsa
def regla_falsa(a, b, tol=1e-4, max_iter=1000):
    fa = f(a)
    fb = f(b)

    # Verificar si hay cambio de signo
    if fa * fb > 0:
        print("No hay cambio de signo en el intervalo, prueba otro intervalo.")
        return None

    # Realizar iteraciones del método de regla falsa
    iter_count = 0
    while (b - a) / 2 > tol:
        c = a - fa * (b - a) / (fb - fa)
        fc = f(c)

        # Si encontramos la raíz exacta
        if fc == 0:
            return c  # La raíz exacta

        # Si la raíz está en el intervalo [a, c]
        elif fa * fc < 0:
            b = c
            fb = fc
        # Si la raíz está en el intervalo [c, b]
        else:
            a = c
            fa = fc

        iter_count += 1
        if iter_count > max_iter:
            print("Se alcanzó el máximo número de iteraciones sin encontrar la raíz.")
            return None

    # Retornar la raíz aproximada
    return c

# a) Graficar la función en el intervalo [-1.5, 1.5]
x_vals = np.linspace(-1.5, 1.5, 400)  # Genera puntos en el intervalo [-1.5, 1.5]
y_vals = f(x_vals)  # Evalúa la función en esos puntos

# Graficar la función
plt.plot(x_vals, y_vals, label=r"$y = 2e^x - 7x^3 - 2$")
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal en y=0
plt.axvline(0, color='black',linewidth=1)  # Línea vertical en x=0
plt.title(r"Gráfica de $y = 2e^x - 7x^3 - 2$")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# d) Encontrar la raíz en el intervalo [-0.7, -0.3]
raiz_1 = regla_falsa(-0.7, -0.3, tol=1e-4)
if raiz_1 is not None:
    print(f"La raíz aproximada en el intervalo [-0.7, -0.3] es: {raiz_1}")

# e) Encontrar la raíz en el intervalo [-0.2, 0.2]
raiz_2 = regla_falsa(-0.2, 0.2, tol=1e-4)
if raiz_2 is not None:
    print(f"La raíz aproximada en el intervalo [-0.2, 0.2] es: {raiz_2}")

# f) Encontrar la raíz en el intervalo [0.5, 0.7]
raiz_3 = regla_falsa(0.5, 0.7, tol=1e-4)
if raiz_3 is not None:
    print(f"La raíz aproximada en el intervalo [0.5, 0.7] es: {raiz_3}")
```
