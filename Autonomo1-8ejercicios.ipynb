# Ejercicio 1


```
import numpy as np
import matplotlib.pyplot as plt
# 1
# Función dada
def f(x):
    return np.tan(x**2) - 1/2

# Método de bisección
def biseccion(a, b, tol=1e-4):
    print(f"f(a) = {f(a)}, f(b) = {f(b)}")  # Imprimir los valores en los extremos
    if f(a) * f(b) > 0:
        print("No hay cambio de signo en el intervalo, prueba otro intervalo.")
        return None

    # Aplicar el método de bisección
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c  # La raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Intervalo adecuado [1.5, 2] (donde la función cambia de signo)
raiz = biseccion(1.5, 2, tol=1e-4)
print(f"La raíz aproximada en el intervalo [1.5, 2] es: {raiz}")

# Graficar la función en el intervalo [0, 3]
x_vals = np.linspace(0, 3, 400)  # Genera puntos en el intervalo [0, 3]
y_vals = f(x_vals)  # Evalúa la función en esos puntos

# Graficar la función
plt.plot(x_vals, y_vals, label=r"$y = \tan(x^2) - \frac{1}{2}$")  # Corregido el formato de LaTeX
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal en y=0
plt.axvline(0, color='black',linewidth=1)  # Línea vertical en x=0
plt.title(r"Gráfica de $y = \tan(x^2) - \frac{1}{2}$")  # Corregido el título con LaTeX
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
# 5
# Función dada
def f(x):
    return 2 * np.exp(x) - 7 * x**3 - 2

# Método de Regla Falsa
def regla_falsa(a, b, tol=1e-4, max_iter=1000):
    fa = f(a)
    fb = f(b)

    # Verificar si hay cambio de signo
    if fa * fb > 0:
        print("No hay cambio de signo en el intervalo, prueba otro intervalo.")
        return None

    # Realizar iteraciones del método de regla falsa
    iter_count = 0
    while (b - a) / 2 > tol:
        c = a - fa * (b - a) / (fb - fa)
        fc = f(c)

        # Si encontramos la raíz exacta
        if fc == 0:
            return c  # La raíz exacta

        # Si la raíz está en el intervalo [a, c]
        elif fa * fc < 0:
            b = c
            fb = fc
        # Si la raíz está en el intervalo [c, b]
        else:
            a = c
            fa = fc

        iter_count += 1
        if iter_count > max_iter:
            print("Se alcanzó el máximo número de iteraciones sin encontrar la raíz.")
            return None

    # Retornar la raíz aproximada
    return c

# a) Graficar la función en el intervalo [-1.5, 1.5]
x_vals = np.linspace(-1.5, 1.5, 400)  # Genera puntos en el intervalo [-1.5, 1.5]
y_vals = f(x_vals)  # Evalúa la función en esos puntos

# Graficar la función
plt.plot(x_vals, y_vals, label=r"$y = 2e^x - 7x^3 - 2$")
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal en y=0
plt.axvline(0, color='black',linewidth=1)  # Línea vertical en x=0
plt.title(r"Gráfica de $y = 2e^x - 7x^3 - 2$")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# d) Encontrar la raíz en el intervalo [-0.7, -0.3]
raiz_1 = regla_falsa(-0.7, -0.3, tol=1e-4)
if raiz_1 is not None:
    print(f"La raíz aproximada en el intervalo [-0.7, -0.3] es: {raiz_1}")

# e) Encontrar la raíz en el intervalo [-0.2, 0.2]
raiz_2 = regla_falsa(-0.2, 0.2, tol=1e-4)
if raiz_2 is not None:
    print(f"La raíz aproximada en el intervalo [-0.2, 0.2] es: {raiz_2}")

# f) Encontrar la raíz en el intervalo [0.5, 0.7]
raiz_3 = regla_falsa(0.5, 0.7, tol=1e-4)
if raiz_3 is not None:
    print(f"La raíz aproximada en el intervalo [0.5, 0.7] es: {raiz_3}")
```

```
# =============================================
# NEWTON–RAPHSON (todo en un solo archivo)
# Problema (preguntas incluidas como comentarios):
#
# Newton-Raphson
# 1.- Sea la función y = 9x^2 + 8x – 2:
#   a) ¿Cuál es el programa para graficar la función y = 9x^2 + 8x – 2 en un rango de –1.5 ≤ x ≤ 0.5?
#   b) ¿Cuál es la gráfica de la función y = 9x^2 + 8x – 2? A simple vista determinar algunos valores aproximados a la raíz x_{n-1}.
#   c) ¿Cuál es el programa para encontrar la raíz de la función y = 9x^2 + 8x – 2, usando para ello el método de Newton-Raphson?
#   d) ¿Cuál es la salida para un valor aproximado de la raíz x_{n-1} = –1.5?
#   e) ¿Cuál es la salida para un valor aproximado de la raíz x_{n-1} = 0.5?
# =============================================

import numpy as np
import matplotlib.pyplot as plt

# ---------- Función y derivada ----------
def f(x):      # y = 9x^2 + 8x - 2
    return 9*x**2 + 8*x - 2

def df(x):     # y' = 18x + 8
    return 18*x + 8

# ---------- (a) y (b) Gráfico en [-1.5, 0.5] ----------
x_min, x_max = -1.5, 0.5
xs = np.linspace(x_min, x_max, 400)
ys = f(xs)

plt.figure()
plt.plot(xs, ys, label="y = 9x^2 + 8x - 2")
plt.axhline(0)                 # eje x
plt.axvline(0, linestyle="--") # eje y
plt.title("Gráfico de y = 9x^2 + 8x - 2  en [-1.5, 0.5]")
plt.xlabel("x"); plt.ylabel("y"); plt.legend()
plt.show()

# Estimación visual rápida (intervalos donde f cambia de signo)
grid = np.linspace(x_min, x_max, 41)  # paso 0.05
vals = f(grid)
print("Posibles intervalos con cambio de signo (estimación visual):")
for i in range(len(grid)-1):
    if vals[i] == 0 or vals[i]*vals[i+1] < 0:
        print(f"  entre x≈{grid[i]:.2f} y x≈{grid[i+1]:.2f}")

# ---------- (c) Método de Newton-Raphson ----------
def newton(x0, tol=1e-6, max_iter=50):
    """
    x_{n} = x_{n-1} - f(x_{n-1})/f'(x_{n-1})
    Devuelve (raiz_aprox, historial) con historial = [(iter, x, f(x), error), ...]
    """
    history = []
    for k in range(1, max_iter+1):
        fx, dfx = f(x0), df(x0)
        if dfx == 0:
            history.append((k, x0, fx, None))
            break
        x1 = x0 - fx/dfx
        err = abs(x1 - x0)
        history.append((k, x1, f(x1), err))
        if err < tol:
            return x1, history
        x0 = x1
    return x0, history

# ---------- (d) Newton con x0 = -1.5 ----------
root_m15, hist_m15 = newton(-1.5)
print("\n(c-d) Método de Newton con x0 = -1.5")
for k, xk, fxk, err in hist_m15:
    print(f"iter {k:2d}: x = {xk:.10f}, f(x) = {fxk:.3e}, error = {0 if err is None else err:.3e}")
print(f"≈ raíz (x0=-1.5): {root_m15:.10f}")

# ---------- (e) Newton con x0 = 0.5 ----------
root_p05, hist_p05 = newton(0.5)
print("\n(e) Método de Newton con x0 = 0.5")
for k, xk, fxk, err in hist_p05:
    print(f"iter {k:2d}: x = {xk:.10f}, f(x) = {fxk:.3e}, error = {0 if err is None else err:.3e}")
print(f"≈ raíz (x0=0.5): {root_p05:.10f}")

# ---------- Comprobación opcional (fórmula cuadrática) ----------
a, b, c = 9, 8, -2
disc = b*b - 4*a*c
x1 = (-b + np.sqrt(disc))/(2*a)
x2 = (-b - np.sqrt(disc))/(2*a)
print("\nComprobación (fórmula cuadrática):")
print(f"x1 = {x1:.10f}, x2 = {x2:.10f}")
```

```
# -------------------------------------------------------
# NEWTON–RAPHSON (Ejercicio 5) — y = tan(x) - x
# a) Graficar y = tan(x) - x
# b) Gráfico en [-2, 2]
# c) Newton-Raphson para hallar raíces de tan(x) - x
# d) Ejecutar con x0 = -1.2
# e) Ejecutar con x0 = 1.2  (comentar por qué puede ir mal)
# f) Ejecutar con x0 = 0
# -------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt

# --- función y derivada ---
def f(x):   return np.tan(x) - x
def df(x):  return 1/np.cos(x)**2 - 1   # = tan(x)^2

# --- (a,b) gráfico en [-2,2] evitando las asíntotas (±pi/2) ---
eps = 1e-3
parts = [(-2, -np.pi/2 - eps), (-np.pi/2 + eps, np.pi/2 - eps), (np.pi/2 + eps, 2)]
plt.figure()
for a,b in parts:
    xs = np.linspace(a, b, 500)
    plt.plot(xs, f(xs))
plt.axhline(0); plt.axvline(0, ls="--")
plt.title("y = tan(x) - x  (rango [-2, 2])")
plt.xlabel("x"); plt.ylabel("y")
plt.show()

# --- (c) Newton-Raphson básico con salvaguardas ---
def newton(x0, tol=1e-6, max_iter=50):
    x = float(x0)
    for k in range(1, max_iter+1):
        # cerca de asíntota o derivada ~ 0 -> detener
        if abs(np.cos(x)) < 1e-6 or abs(df(x)) < 1e-12:
            return x, k, "falló: derivada ~0 o cerca de asíntota"
        x1 = x - f(x)/df(x)
        if abs(x1 - x) < tol:
            return x1, k, "ok"
        x = x1
    return x, max_iter, "máx. iteraciones"

# (d) x0 = -1.2
r1, it1, msg1 = newton(-1.2, 1e-6)
print(f"(d) x0=-1.2  -> raíz≈ {r1:.10f}, it={it1}, estado={msg1}")

# (e) x0 = 1.2
r2, it2, msg2 = newton(1.2, 1e-6)
print(f"(e) x0= 1.2  -> raíz≈ {r2:.10f}, it={it2}, estado={msg2}")
print("Nota (e): tan(x) tiene asíntota en x=+π/2≈1.5708, cerca del x0=1.2.\n"
      "La pendiente puede volverse muy grande o la derivada casi nula, "
      "lo que hace que Newton salte o converja lentamente; por eso el valor "
      "puede no ser muy cercano a la raíz real.")

# (f) x0 = 0
if abs(f(0.0)) < 1e-12:
    print("(f) x0=0  -> ya estamos en la raíz exacta x=0,")
    print("     pero Newton no puede dar el siguiente paso porque f'(0)=0 (división por cero).")
else:
    r3, it3, msg3 = newton(0.0, 1e-6)
    print(f"(f) x0=0   -> raíz≈ {r3:.10f}, it={it3}, estado={msg3}")
```

# Ejercicio 6

```
# Problema 6 – Método de la Secante para f(x) = 5 sin(x) − x^3
# a,b) Graficar en [-2, 2]
# c) Implementar secante
# d) a=-2, b=-1.5, tol=1e-4
# e) a=-0.5, b=0.5, tol=1e-4
# f) a=1.6, b=1.9, tol=1e-4
# g) a=1.0, b=1.5, tol=1e-4  + explicación

import numpy as np
import matplotlib.pyplot as plt

# --------- función ----------
def f(x):
    return 5*np.sin(x) - x**3

# --------- método de la secante ----------
def secant(a, b, tol=1e-4, max_iter=100):
    """
    Retorna (converged, x, history, message)
    history: lista de tuplas (k, x_{k-1}, x_k, f(x_{k-1}), f(x_k), |x_k - x_{k-1}|)
    Criterio: |x_{k+1} - x_k| <= tol
    Maneja divisiones por cero cuando f(x_k)-f(x_{k-1})≈0.
    """
    x0, x1 = float(a), float(b)
    f0, f1 = f(x0), f(x1)
    hist = []
    for k in range(1, max_iter + 1):
        denom = (f1 - f0)
        if abs(denom) < 1e-14 or not np.isfinite(denom):
            return False, x1, hist, "Denominador ~0 en la fórmula de la secante."
        x2 = x1 - f1*(x1 - x0)/denom
        delta = abs(x2 - x1)
        hist.append((k, x0, x1, f0, f1, delta))
        if delta <= tol:
            return True, x2, hist, "OK"
        x0, x1 = x1, x2
        f0, f1 = f1, f(x1)
    return False, x1, hist, "Se alcanzó max_iter"

# --------- (a) y (b) Gráfica ----------
xs = np.linspace(-2, 2, 1000)
ys = f(xs)
plt.axhline(0, linewidth=1)
plt.plot(xs, ys, label="y = 5 sin(x) − x^3")
plt.xlabel("x"); plt.ylabel("y"); plt.title("Gráfica en [-2, 2]")
plt.legend(); plt.show()

# --------- Utilidad para imprimir tablas ----------
def print_table(title, result):
    conv, root, hist, msg = result
    print("\n", title)
    if conv:
        print(f"Convergió a x ≈ {root:.12f} en {len(hist)} iteraciones")
    else:
        print(f"No convergió (último x ≈ {root:.12f}). Motivo: {msg}")
    print("k\t   x_{k-1}\t\t      x_k\t\t    f(x_{k-1})\t\t      f(x_k)\t\t      |Δx|")
    for k, xkm1, xk, fxkm1, fxk, dx in hist:
        print(f"{k}\t {xkm1:.10f}\t {xk:.10f}\t {fxkm1:.10e}\t {fxk:.10e}\t {dx:.6e}")

# --------- (d) a=-2, b=-1.5 ---------
res_d = secant(a=-2.0, b=-1.5, tol=1e-4)
print_table("d) a=-2, b=-1.5, tol=1e-4", res_d)

# --------- (e) a=-0.5, b=0.5 ---------
res_e = secant(a=-0.5, b=0.5, tol=1e-4)
print_table("e) a=-0.5, b=0.5, tol=1e-4", res_e)

# --------- (f) a=1.6, b=1.9 ---------
res_f = secant(a=1.6, b=1.9, tol=1e-4)
print_table("f) a=1.6, b=1.9, tol=1e-4", res_f)

# --------- (g) a=1.0, b=1.5 ---------
res_g = secant(a=1.0, b=1.5, tol=1e-4)
print_table("g) a=1.0, b=1.5, tol=1e-4", res_g)

# --------- Explicación (g):
"""
La secante es un método *abierto*: no garantiza permanecer dentro del intervalo
[a,b] como sí lo hacen los métodos de bisección o falsa posición. Cada nuevo
x_{k+1} se calcula con la intersección de la recta secante por (x_{k-1}, f(x_{k-1}))
y (x_k, f(x_k)), y ese punto puede caer fuera de [a,b]. Si la función es suave
y el par inicial está en la cuenca de atracción de otra raíz (por la curvatura
de f y las pendientes en los puntos iniciales), la iteración puede “saltar”
hacia esa raíz y converger fuera del intervalo original [1, 1.5].
"""
```

```
# Problema 9 – Método de la secante
# f(x) = sqrt(5x^2 + 6x) - x^3,  x >= 0
# a,b) Graficar en [0, 2.5]
# c) Secante con tolerancia |x_{k+1}-x_k| <= 1e-4
# d) a=1.8, b=2.0
# e) a=0.25, b=0.5  (discusión de no convergencia hacia x=0)

import numpy as np
import matplotlib.pyplot as plt

# ---------- función ----------
def f(x: float) -> float:
    x = np.asarray(x)
    return np.sqrt(5*x**2 + 6*x) - x**3  # dominio x >= 0

# ---------- método de la secante ----------
def secante(a, b, tol=1e-4, max_iter=100):
    """
    Devuelve (converged, x_aprox, history, msg)
    history = [(k, x_{k-1}, x_k, f(x_{k-1}), f(x_k), |x_k-x_{k-1}|), ...]
    """
    x0, x1 = float(a), float(b)
    if x0 < 0 or x1 < 0:
        raise ValueError("Los puntos iniciales deben cumplir x >= 0 por el radical.")
    f0, f1 = f(x0), f(x1)
    hist = []
    for k in range(1, max_iter + 1):
        den = f1 - f0
        if not np.isfinite(den) or abs(den) < 1e-14:
            return False, x1, hist, "Denominador ~0 en la fórmula de la secante"
        x2 = x1 - f1*(x1 - x0)/den
        dx = abs(x2 - x1)
        hist.append((k, x0, x1, f0, f1, dx))
        if dx <= tol:
            return True, x2, hist, "OK"
        x0, x1 = x1, x2
        if x1 < 0:  # fuera del dominio
            return False, x1, hist, "Iteración salió del dominio (x < 0)."
        f0, f1 = f1, f(x1)
    return False, x1, hist, "Se alcanzó max_iter"

def imprimir_tabla(titulo, res):
    conv, root, hist, msg = res
    print(f"\n{titulo}")
    if conv:
        print(f"Convergió a x ≈ {root:.12f} en {len(hist)} iteraciones")
    else:
        print(f"No convergió. Último x ≈ {root:.12f}. Motivo: {msg}")
    print("k\t   x_{k-1}\t\t      x_k\t\t    f(x_{k-1})\t\t      f(x_k)\t\t      |Δx|")
    for k, xkm1, xk, fxkm1, fxk, dx in hist:
        print(f"{k}\t {xkm1:.10f}\t {xk:.10f}\t {fxkm1:.10e}\t {fxk:.10e}\t {dx:.6e}")

# ---------- (a) y (b) Gráfica ----------
xs = np.linspace(0, 2.5, 600)
ys = f(xs)
plt.axhline(0, linewidth=1)
plt.plot(xs, ys, label=r"$f(x)=\sqrt{5x^2+6x}-x^3$")
plt.xlabel("x"); plt.ylabel("y"); plt.title("Gráfica en [0, 2.5]")
plt.legend(); plt.show()

# ---------- (d) a=1.8, b=2.0 ----------
res_d = secante(1.8, 2.0, tol=1e-4)
imprimir_tabla("d) Secante con a=1.8, b=2.0, tol=1e-4", res_d)

# ---------- (e) a=0.25, b=0.5 ----------
res_e = secante(0.25, 0.5, tol=1e-4)
imprimir_tabla("e) Secante con a=0.25, b=0.5, tol=1e-4", res_e)

# Comentario para (e):
"""
Cerca de x=0, la parte radical domina: sqrt(5x^2+6x) ~ sqrt(6x).
Su derivada es ~ 3/sqrt(6x), que → ∞ cuando x → 0^+.
Esto produce una pendiente extremadamente grande y una curvatura fuerte
en las proximidades de la raíz x=0. Con puntos iniciales a=0.25 y b=0.5,
f(a) y f(b) son positivos y muy parecidos; la secante resultante corta el eje
x lejos del 0 (incluso puede saltar fuera del dominio x>=0), por lo que
la iteración puede alejarse y **no converger** a la raíz x=0.
La secante es un método *abierto* y no garantiza permanecer en [a,b].
"""
```

```
import cmath

def f(x):
    return x**4 - 3*x**3 + x**2 + x + 1

def muller(x0, x1, x2, tol=1e-5, maxit=20):
    for i in range(1, maxit+1):
        f0, f1, f2 = f(x0), f(x1), f(x2)
        h0 = x1 - x0
        h1 = x2 - x1
        d0 = (f1 - f0)/h0
        d1 = (f2 - f1)/h1
        a = (d1 - d0)/(h1 + h0)
        b = a*h1 + d1
        c = f2
        rad = cmath.sqrt(b*b - 4*a*c)
        if abs(b+rad) > abs(b-rad):
            den = b+rad
        else:
            den = b-rad
        dx = -2*c/den
        x3 = x2 + dx
        print(i+2, x3, f(x3))   # aquí ves cada iteración
        if abs(dx) < tol:
            break
        x0, x1, x2 = x1, x2, x3

print("Raíz compleja")
muller(0.5, -0.5, 0)

print("\nRaíz real pequeña")
muller(0.5, 1.0, 1.5)

print("\nRaíz real grande")
muller(1.5, 2.0, 2.5)
```
